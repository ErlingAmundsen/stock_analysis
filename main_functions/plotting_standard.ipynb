# Plotting data

Made a standardized cool method to plot data, that is also interactable (using plotly)

```python
import plotting_standard as ps
import format_yahoo as fy
import plotly.graph_objects as ply

data, meta = fy.format_yahoo('BTC-USD', period= '1y')
fig = ps.candle_plot(data=data, meta=meta)


# example line we might want to add
data['line'] = [18000 + i*50 for i in range(len(data))]
# proving that it can work on smaller parts of the data
data['line'][:50] = None

# How we would add it
fig.add_trace(ply.Scatter(
    x = data.index,
    y = data['line'],
    mode = 'lines',
    name= 'added line'
)
)

fig.show()
```
